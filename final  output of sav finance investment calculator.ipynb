```python
#login page

from colorama import Fore,Back,Style 

from IPython.display import Image

print("-"*46,"SAV FINANCE INVESTMENT CALCULATOR","-"*46) 

display(Image(r'C:\\Users\\Akshaya\\Pictures\\sav 9.jpeg'))

print(Fore.CYAN+("~"*127)) 

print(Fore.YELLOW+"[1]- NEW USER") 

print(Fore.GREEN+"[2]-ALREADY AN EXSISTING USER ") 

print() 

print(Fore.CYAN+("~"*127)) 

ch=int(input("ENTER THE CHOICE :")) 

print(Fore.CYAN+("~"*127)) 

if ch==1: 

    print() 

    print(Fore.YELLOW+"NEW USER") 

    name=input("ENTER YOUR NAME:") 

    phone=int(input("ENTER YOUR PHONE NUMBER:")) 

    password=input("CREATE A PASSWORD:") 

elif ch==2: 

    print() 

    print(Fore.GREEN+"ALREADY AN EXISTING USER") 

    name=input("ENTER YOUR NAME:") 

    password=input("ENTER YOUR PASSWORD:") 

else: 

    print("INVALID CHOICE") 

print() 

print(Fore.MAGENTA+">"*50,"SUCCESSFULLY LOGGED IN ","<"*50) 

print() 

print(Fore.MAGENTA+" "*35,"WELCOME TO SAV FINANCE INVESTMENT CALCULATOR") 

#showing applications  

print() 

print(Fore.YELLOW+"[1]-TO CALCULATE THE SIMPLE INTEREST") 

print(Fore.GREEN+"[2]-TO CALCULATE THE COMPOUND INTEREST") 

print(Fore.BLUE+"[3]-TO CALCULATE THE GST") 

print(Fore.RED+"[4]-TO CALCULATE THE EMI") 

print(Fore.MAGENTA+"[5]-TO CALCULATE THE FIXED DEPOSIT") 

print() 

application=int(input("ENTER THE APPLICATION TO USE:")) 

print() 

if application==1: 

    print(Fore.YELLOW+"SIMPLE INTEREST") 

    principal_amount=float(input("ENTER THE PRINCIPAL AMOUNT:")) 

    rate_interest=float(input("ENTER THE RATE OF INTEREST:")) 

    no_years=int(input("ENTER THE NUMBER OF THE YEARS:")) 

    print() 

    si=(principal_amount*rate_interest*no_years)/100 

    print("THE SIMPLE INTEREST IS:",si) 

elif application==2: 

    print(Fore.GREEN+"COMPOUND INTEREST") 

    prin=float(input("ENTER THE PRINCIPAL AMOUNT:")) 

    rate=int(input("ENTER THE RATE OF INTEREST:")) 

    no=int(input("ENTER THE NUMBER OF THE YEARS:")) 

    print() 

    amount=prin*((1+rate/100)**no) 

    compound=amount-prin 

    print("THE AMOUNT IS:",amount) 

    print("THE COMPOUND INTEREST IS:",compound) 

elif application==3: 

    print(Fore.CYAN+"GST") 

    op=int(input("ENTER THE ORIGINAL PRICE:")) 

    gstrate=int(input("ENTER THE RATE OF GST:")) 

    print() 

    cal=(op*gstrate)/100 

    total=op+cal 

    print("TOTAL GST:",total) 

elif application==4: 

    print(Fore.RED+" MONTHLY INSTALMENT (EMI)") 

    prin=float(input("ENTER THE PRINCIPAL AMOUNT:")) 

    rate=int(input("ENTER THE RATE OF INTEREST:")) 

    no=int(input("ENTER THE NUMBER OF THE YEARS:")) 

    print() 

    monthly_int_rate=rate/(12*100) 

    month=no*12; 

    emi=prin*monthly_int_rate*((1+monthly_int_rate)**month)/((1+monthly_int_rate)**month-1) 

    print("YOUR MONTHLY INSTALLMENT(EMI):",emi) 

elif application ==5: 

    print(Fore.MAGENTA+"FIXED DEPOSIT") 

    prin=float(input("ENTER THE PRINCIPAL AMOUNT:")) 

    rate=int(input("ENTER THE RATE OF INTEREST:")) 

    no=int(input("ENTER THE NUMBER OF THE YEARS:")) 

    print() 

    interest=(prin*rate*no)/100 

    amount=prin+interest 

    print("THE INTEREST IS :",interest) 

    print("THE AMOUNT IS:",amount) 

else: 

    print(Fore.RED+"INVALID OPTION ") 

    print(Fore.RED+"ENTER A VALID OPTION") 

    print(Fore.CYAN+"~"*127) 
    
print()

print("DO YOU WANT TO EXIT")

print()

print(Fore.BLUE+"[1]-YES") 

print()

print(Fore.GREEN+"[2]-NO") 

print()

choice=int(input("ENTER YOUR CHOICE:")) 

if choice==1: 

    print(Fore.BLUE+"YES") 

    print(Fore.MAGENTA+">"*58,"THANK YOU","<"*58) 

elif choice==2: 

    print(Fore.GREEN+"NO,I WANT TO EXIST")
```
